# MovieLens 1M Dataset

```python
import pandas as pd
```

```python
users = pd.read_table(
    'datasets/movielens/users.dat',
    names= ['user_id', 'gender', 'age', 'occupation', 'zip_code'],
    header= None,
    delimiter= '::',
    engine= 'python'
)
```

```python
users.head()
```

```python
movies = pd.read_table(
    'datasets/movielens/movies.dat',
    names= ['movie_id', 'title', 'genres'],
    header= None,
    delimiter= '::',
    engine= 'python'
)
```

```python
movies.head()
```

```python
ratings = pd.read_table(
    'datasets/movielens/ratings.dat',
    names= ['user_id', 'movie_id', 'rating', 'timestamp'],
    header= None,
    delimiter= '::',
    engine= 'python'
)
```

```python
ratings.head()
```

```python
merged_data = pd.merge(pd.merge(ratings, users), movies)
```

```python
merged_data
```

```python
merged_data.info()
```

## Finding mean rating of movies (gender wise)

```python
mean_ratings = merged_data.pivot_table(
    'rating',
    index= 'title',
    columns= 'gender',
    aggfunc= 'mean'
)
```

```python
mean_ratings.head()
```

## Movies with at least 250 ratings

```python
ratings_by_title = merged_data.groupby('title').size()
```

```python
ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
```

## Mean rating of active titles (gender wise)

```python
mean_ratings = mean_ratings.loc[active_titles]
```

```python
mean_ratings.head()
```

## Top female ratings

```python
top_female_ratings = mean_ratings.sort_values('F', ascending= False)
```

```python
top_female_ratings.head()
```

## Measuring Rating Disagreement

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
mean_ratings.head()
```

```python
sorted_by_diff = mean_ratings.sort_values(by= 'diff')
```

```python
sorted_by_diff.head()
```

```python
rating_std_by_title = merged_data.groupby('title')['rating'].std()
```

```python
rating_std_by_title = rating_std_by_title.loc[active_titles]
```

```python
rating_std_by_title.head()
```

```python
rating_std_by_title.sort_values(ascending= False).head(10)
```

## Splitting genres and exploding

```python
movies.genres.head()
```

```python
splitted_genres = movies.genres.str.split('|')
```

```python
splitted_genres.head()
```

```python
movies.genres = splitted_genres
```

```python
movies.head()
```

```python
exploded_movies = movies.explode('genres')
```

```python
exploded_movies.head(10)
```

## Age wise genre ratings

```python
ratings_with_genres = pd.merge(pd.merge(exploded_movies, ratings), users)
```

```python
ratings_with_genres.iloc[0]
```

```python
genre_ratings = (
    ratings_with_genres.groupby(['genres', 'age'])
    ['rating']
    .mean()
    .unstack('age')
)
```
