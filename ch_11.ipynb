## Grouped Time Resampling

```python
N = 15
```

```python
times = pd.date_range('2017-05-20 00:00:00', freq= '1T', periods= N)
```

```python
times
```

```python
df = pd.DataFrame(
    {
        'time': times,
        'value': np.arange(N)
    }
)
```

```python
df.head()
```

```python
df.set_index('time').resample('5T').sum()
```

```python
df2 = pd.DataFrame(
    {
        'time': times.repeat(3),
        'key': np.tile(['a', 'b', 'c'], N),
        'value': np.arange(N * 3)
    }
)
```

```python
df2.head()
```

```python
time_key = pd.Grouper(freq= '5min')
```

```python
time_key
```

```python
resampled = (
    df2.set_index('time')
    .groupby(['key', time_key])
    .sum()
)
```

```python
resampled
```

```python
resampled.reset_index()
```

## Exponentially Weighted Functions

```python
aapl = close_px.loc['2006':'2007', 'AAPL']
```

```python
aapl
```

```python
ma30 = aapl.rolling(30, min_periods= 20).mean()
```

```python
ma30
```

```python
ewm30 = aapl.ewm(span= 30).mean()
```

```python
ewm30
```

```python
aapl.plot(label= 'Price')
ma30.plot(label= 'Simple Moving Average')
ewm30.plot(label= 'Exponentially Weighted', style= '--')
plt.legend()
plt.title('Simple moving average VS Exponentially weighted moving average')
```

## User-Defined Moving Window Functions

```python
from scipy.stats import percentileofscore
```

```python
def score_at_2_percent(x):
    return percentileofscore(x, 0.02)
```

```python
result = returns['AAPL'].rolling(250).apply(score_at_2_percent)
```

```python
result.plot()
plt.title('Percentile rank of 2% AAPL return over one-year window')
```
