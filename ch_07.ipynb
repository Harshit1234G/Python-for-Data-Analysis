## Computations with Categoricals

```python
draws = np.random.standard_normal(1000)
draws[:5]
```

```python
bins = pd.qcut(draws, 4)
```

```python
bins
```

```python
bins = pd.qcut(draws, 4, labels= ['Q1', 'Q2', 'Q3', 'Q4'])
```

```python
bins
```

```python
bins = pd.Series(bins, name= 'quartile')
```

```python
bins
```

```python
result = (
    pd.Series(draws)
    .groupby(bins)
    .agg(['min', 'max', 'count'])
    .reset_index()
)
```

```python
result
```
