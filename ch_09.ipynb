## Figures and Subplots

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax1.hist(np.random.standard_normal(100), bins= 20, color= 'red', alpha= .3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
ax3.plot(np.random.standard_normal(50).cumsum(), color= 'black', linestyle= 'dashed')
```

```python
fig, axes = plt.subplots(2, 3)
axes[0, 1].plot(np.random.standard_normal(30))
```

```python
axes
```

### Adjusting the spacing around subplots

```python
fig, axes = plt.subplots(2, 2, sharex= True, sharey= True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins= 50, color= 'black', alpha= .3)
fig.subplots_adjust(wspace= 0, hspace= 0)
```

## Ticks, Labels, and Legends
### Setting the title, axis labels, ticks and tick labels

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation= 30, fontsize= 8)
ax.set_xlabel('stages')
ax.set_title('Random Plot')
```

### Adding legends

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum(), label= 'one')
ax.plot(np.random.standard_normal(1000).cumsum(), color= 'green', label= 'two')
ax.plot(np.random.standard_normal(1000).cumsum(), color= 'red', label= 'three')
ax.legend()
ax.set_title('Three plots with legend')
```

## Annotations and Drawing on Subplot

```python
from datetime import datetime
```

```python
data = pd.read_csv('examples/spx.csv', index_col= 0, parse_dates= True)
```

```python
data
```

```python
spx = data.SPX
```

```python
spx
```

```python
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
```

```python
fig, ax = plt.subplots()
ax.plot(spx)

for date, label in crisis_data:
    ax.annotate(
        text= label,
        xy= (date, spx[date] + 50),
        xytext= (date, spx[date] + 200),
        arrowprops= {
            'facecolor': 'black',
            'headwidth': 5,
            'width': 2,
            'headlength': 4
        },
        horizontalalignment= 'left',
        verticalalignment= 'top'
    )

ax.set_xlim([datetime(2007, 1, 1), datetime(2011, 1, 1)])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2007-2009 financial crisis')
ax.set_xlabel('Date')
plt.xticks(rotation= 45)
```

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color= 'red', alpha= 0.5)
circle = plt.Circle((0.7, 0.2), 0.15, color= 'green', alpha= 0.5)
polygon = plt.Polygon(
    [
        [0.15, 0.15],
        [0.35, 0.4],
        [0.2, 0.6]
    ],
    color= 'blue',
    alpha= 0.5
)

ax.add_patch(rect)
ax.add_patch(circle)
ax.add_patch(polygon)
```

## Facet Grids and Categorical Data

```python
tips.head()
```

```python
sns.catplot(
    data= tips.loc[tips.tip_percentage < 1],
    x= 'day',
    y= 'tip_percentage',
    col= 'smoker',
    hue= 'time',
    kind= 'bar'
)
```

```python
sns.catplot(
    data= tips.loc[tips.tip_percentage < 1],
    x= 'day',
    y= 'tip_percentage',
    col= 'smoker',
    row= 'time',
    kind= 'bar'
)
```

```python
sns.catplot(
    data= tips.loc[tips.tip_percentage < 0.5],
    x= 'day',
    y= 'tip_percentage',
    kind= 'box'
)
```

```python
sns.boxplot(
    data= tips.loc[tips.tip_percentage < 0.5],
    x= 'tip_percentage',
    y= 'day'
)
```
