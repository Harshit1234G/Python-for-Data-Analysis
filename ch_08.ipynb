# Combining and Merging Datasets

```python
df1 = pd.DataFrame(
    {
        "key": ["b", "b", "a", "c", "a", "a", "b"],
        "data1": pd.Series(range(7), dtype="Int64")
    }
)
df2 = pd.DataFrame(
    {
        "key": ["a", "b", "d"],
        "data2": pd.Series(range(3), dtype="Int64")
    }
)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on= 'key')
```

```python
df3 = pd.DataFrame(
    {
        "lkey": ["b", "b", "a", "c", "a", "a", "b"],
        "data1": pd.Series(range(7), dtype="Int64")
    }
)
```

```python
df3
```

```python
df4 = pd.DataFrame(
    {
        "rkey": ["a", "b", "d"],
        "data2": pd.Series(range(3), dtype="Int64")
    }
)
```

```python
df4
```

```python
pd.merge(df3, df4, left_on= 'lkey', right_on= 'rkey')
```

```python
pd.merge(df1, df2, how= 'outer')
```

```python
pd.merge(df3, df4, how= 'outer', left_on= 'lkey', right_on= 'rkey')
```

```python
pd.merge(df1, df2, how= 'left')
```

```python
pd.merge(df1, df2, how= 'right')
```

```python
left = pd.DataFrame(
    {
        "key1": ["foo", "foo", "bar"],
        "key2": ["one", "two", "one"],
        "lval": pd.Series([1, 2, 3], dtype='Int64')
    }
)
```

```python
right = pd.DataFrame(
    {
        "key1": ["foo", "foo", "bar", "bar"],
        "key2": ["one", "one", "one", "two"],
        "rval": pd.Series([4, 5, 6, 7], dtype='Int64')
    }
)
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how= 'outer', on= ['key1', 'key2'])
```

```python
pd.merge(left, right, on= 'key1')
```

```python
pd.merge(left, right, on= 'key1', suffixes= ['_left', '_right'])
```

```python
data
```

```python
result = data.stack()
```

```python
result
```

```python
result.unstack()
```

```python
result.unstack(level= 0)
```

```python
result.unstack(level= 'state')
```

```python
data2
```

## Pivoting "Long" to "Wide" Format

```python
data = pd.read_csv('examples/macrodata.csv')
```

```python
data = data.loc[:, ['year', 'quarter', 'realgdp', 'infl', 'unemp']]
```

```python
data
```

```python
periods = pd.PeriodIndex(
    year= data.pop('year'),
    quarter= data.pop('quarter'),
    name= 'date'
)
```

```python
periods
```

```python
data.index = periods.to_timestamp('D')
```

```python
data
```

```python
data.columns.name = 'item'
```

```python
data.head()
```

```python
data.stack()
```

```python
long_data = data.stack().reset_index().rename(columns= {0: 'value'})
```

```python
long_data
```

```python
pivoted_data = long_data.pivot(index= 'date', columns= 'item', values= 'value')
```

```python
pivoted_data
```

```python
long_data['value2'] = np.random.standard_normal(len(long_data))
```

```python
long_data
```

```python
pivoted = long_data.pivot(index= 'date', columns= 'item')
```

```python
pivoted
```

```python
pivoted.value.head()
```

```python
pivoted.value2.head()
```

```python
long_data.set_index(['date', 'item']).unstack(level= 'item')
```

## Pivoting "Wide" to "Long" Format

```python
df = pd.DataFrame(
    {
        "key": ["foo", "bar", "baz"],
        "A": [1, 2, 3],
        "B": [4, 5, 6],
        "C": [7, 8, 9]
    }
)
```

```python
df
```

```python
pd.melt(df)
```

```python
melted = pd.melt(df, id_vars= 'key')
```

```python
melted
```

```python
reshaped = melted.pivot(index= 'key', columns= 'variable', values= 'value')
```

```python
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df, id_vars= 'key', value_vars= ['A', 'B'])
```

```python
pd.melt(df, id_vars= 'key', value_vars= ['A', 'B', 'C'])
```
