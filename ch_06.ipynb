```python
chunker
```

```python
tot = pd.Series([], dtype= 'int64')
```

```python
for piece in chunker:
  tot = tot.add(piece.key.value_counts(), fill_value= 0)
```

```python
tot = tot.sort_values(ascending= False)
```

## Working with Other Delimited Formats

```python
!cat examples/ex7.csv
```

```python
import csv
```

```python
with open('examples/ex7.csv') as f:
  reader: list[list[str]] = list(csv.reader(f))
```

```python
for line in reader:
  print(line)
```

```python
class my_dialect(csv.Dialect):
    delimiter: str = ';'
    lineterminator: str = '\n'
    quotechar: str = '"'
    quoting = csv.QUOTE_MINIMAL
```

```python
with open('examples/my_data.csv', 'w') as f:
    writer = csv.writer(f, dialect= my_dialect)
    writer.writerow(('one', 'two', 'three'))
    writer.writerow(('1', '2', '3'))
    writer.writerow(('4', '5', '6'))
```

# Interacting with Databases

```python
import sqlite3
```

```python
query = """
    CREATE TABLE student(
        roll_no INTEGER,
        name TEXT
    );
"""
```

```python
connector = sqlite3.connect('data.sqlite')
```

```python
connector.execute(query)
connector.commit()
```

```python
connector.executemany(
    """
    INSERT INTO student(roll_no, name)
    VALUES(?, ?);
    """,
    [
        (1, 'Harshit'),
        (2, 'Kushal'),
        (3, 'Ruchi')
    ]
)
connector.commit()
```

```python
connector.execute(
    """
    SELECT *
    FROM student;
    """
).fetchall()
```

```python
pd.read_sql('SELECT * FROM student', connector)
```
